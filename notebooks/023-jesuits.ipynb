## **[PT]** Listagem

```python
import pandas as pd
from timelinknb.pandas import attribute_to_df

# Get the member of Colégio de Jesus
cjhs = attribute_to_df(
                    the_type='colegio',
                    the_value='Colégio de Jesus',
                    person_info=True,
                    more_cols=['naturalidade','faculdade','uc-entrada','grau'])
cjhs.info()
```

```python
cjhs[['name','naturalidade','faculdade','grau','uc-entrada']].fillna("")
```

```python

# Get the Jesuits
cjesus = attribute_to_df(
                    the_type='ordem-religiosa',
                    the_value='Companhia de Jesus',
                    person_info=True,
                    more_cols=['naturalidade','faculdade','uc-entrada','grau'],
                    sql_echo=False)
cjesus.info()
```

```python
cjesus[['name','naturalidade','faculdade','grau','uc-entrada']].sort_values(['name'])
```

### **[EN]**

Get the ids of both sets

```python
jesuits = set(cjhs.index.values).union(cjesus.index.values) 
```

```python
# get extra Jesuits not indetified as such

more = [
    138650,  # Belchior [ou Melchior] Carneiro Bispo, votos 1543, bispo de Niceia, governador do bispado de Macau
]

jesuits = list(jesuits) + more
```

### **[EN]**

Collective attributes

```python
from timelinknb.pandas import group_attributes
pd.set_option('display.max_rows',250)

exclude = ['uc-entrada.ano','uc-saida.ano','nota-processamento','código-de-referência','grau.ano','faculdade.ano']
include = ['uc-entrada','grau','faculdade','naturalidade','data-do-registo']
jesuits_df = group_attributes(jesuits,person_info=True, include_attributes=include,exclude_attributes=exclude)
jesuits_df['ano']=jesuits_df.date.str[0:4]
lista = jesuits_df[['ano','date','name','type','value','attr_obs']].sort_values(['date','name'])
lista.fillna("")
```
