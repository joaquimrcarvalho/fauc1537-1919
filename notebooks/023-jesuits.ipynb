## Colégio de Jesus

---

## College of Jesus

```python
import pandas as pd
from timelinknb.pandas import attribute_to_df

# Get the member of Colégio de Jesus
cjhs = attribute_to_df(
                    the_type='colegio',
                    the_value='Colégio de Jesus',
                    person_info=True,
                    dates_in=('1500','1950'),
                    more_cols=['naturalidade','faculdade','uc-entrada','grau'])
cjhs.info()
```

```python
cjhs[['name','naturalidade','faculdade','grau','uc-entrada']].fillna("")
```


## Companhia de Jesus

---

## Jesuits as order

```python

# Get the Jesuits
cjesus = attribute_to_df(
                    the_type='ordem-religiosa',
                    the_value='Companhia de Jesus',
                    person_info=True,
                    more_cols=['naturalidade','faculdade','uc-entrada','grau'],
                    dates_in=('1500','1950'),
                    sql_echo=False)
cjesus.info()
```

```python
cjesus[['name','naturalidade','faculdade','grau','uc-entrada']].sort_values(['name']).fillna("")
```

## Juntar ambos

---

## Join both

```python
jesuits = set(cjhs.index.values).union(cjesus.index.values) 
```

## Other Jesuits, not identified as such

```python
# get extra Jesuits not indetified as such

more = [
    '138650',  # Belchior [ou Melchior] Carneiro Bispo, votos 1543, bispo de Niceia, governador do bispado de Macau
    '232178',  # Francisco Suarez
    '141304',  # Miguel Acciaioly
    '134282',  # António José de Figueiredo
    '152879',  # Estevão Dias
    '203261',  # Manuel Américo da Costa (padre)
    '218204',  # Manuel Vieira
    '182670',  # Baltasar de Magalhães
    '211163',  # Luís Leite
    '234839',  # Francisco Martins Pereira (padre)
]

jesuits = list(jesuits) + more
```

### Biografia colectiva

---

### Collective biography

```python
from timelinknb.pandas import group_attributes
pd.set_option('display.max_rows',250)

exclude = ['uc-entrada.ano','uc-saida.ano','nota-processamento','código-de-referência','grau.ano','faculdade.ano','data-do-registo']
include = ['uc-entrada','grau','faculdade','naturalidade','data-do-registo','ordem-religiosa']
jesuits_df = group_attributes(jesuits,person_info=True, include_attributes=include,exclude_attributes=exclude)
jesuits_df['ano']=jesuits_df.date.str[0:4]
lista = jesuits_df[['ano','date','name','type','value','attr_obs']].sort_values(['date','name'])
lista.fillna("")
```
