```python

from timelinknb.pandas import attribute_to_df


# Get list of people with a given attribute type, extra columns
doutores = attribute_to_df(
                    the_type='grau',
                    the_value='doutor%',
                    person_info=True,
                    more_cols=['uc-entrada','uc-saida','faculdade','nota-processamento','naturalidade','faculdade-original'])
doutores.info()
```

**[EN]** Export list

```python
doutores.sort_values(['name','faculdade','uc-entrada']).to_csv('../inferences/academic/doutores.csv',sep=',')
```

**[EN]** Compare with Vasconcelos, António de, Escritos Vários, Reed (Coimbra: Arquivo da Universidade de Coimbra, 1987), v. II, listing from p. 40 e ss.

Note that Vasconcelos only lists dregrees from 1801 onwards.

```python
sec19=doutores[doutores['grau.date'] > '1801-00-00'].sort_values(['grau.date'])
sec19.sort_values(['grau.date']).to_csv('../inferences/validation/doutores_vasconcelos_ev_ii_p.40.csv',sep=',')
```

```python
pd.set_option('display.max_rows',1000)
cols = ['name','faculdade','grau.date','grau','nota-processamento','faculdade-original','uc-entrada','uc-saida','naturalidade']
sec19[cols].sort_values('grau.date').fillna("").head(50)
```
